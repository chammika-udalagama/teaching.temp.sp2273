# 1 A simple plot

```python
from matplotlib import pyplot as plt
import numpy as np
import scienceplots
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .2, 0.75

for alpha in [.25,.75,1]:
    plt.clf()
    plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed')
    plt.plot(x, y_2, label='Y2 values', color='blue', )
    plt.errorbar(x, y_3,  xerr=x_error, yerr=y_error,
                  label='Y3 with errors', color='green')
    plt.xlabel('x-values')
    plt.ylabel('y-values')
    plt.title('X vs Y')
    plt.grid(alpha=alpha)
    plt.legend(loc='upper left')
    plt.savefig(f'hahaha_{alpha=}.png', dpi=300)
    
plt.show()
```

## 4.1 Plotting data from files

```python
file_url='https://berkeley-earth-temperature.s3.us-west-1.amazonaws.com/Global/Complete_TAVG_daily.txt'
data=np.loadtxt(file_url, skiprows=23)
```

```python
years = data[:,1]
anomaly=data[:,-1]
plt.plot(years,anomaly, alpha=.5)

year_min,year_max = years.min(), years.max()
plt.hlines(0,year_min, year_max, color='r', ls='dashed' )
```
